# 📥 Tầng 1: Data Ingestion

**Mục tiêu:** Đọc file raw `CI_CD Survey Output - Student.csv` và lưu thành `responses.csv` (không thay đổi cấu trúc).

**Input:** `CI_CD Survey Output - Student.csv`  
**Output:** `responses.csv`

```python
import pandas as pd
from pathlib import Path

# ── Config ──────────────────────────────────────────────
RAW_FILE = "CI_CD Survey Output - Student.csv"
OUTPUT_FILE = "responses.csv"

# ── Đọc file raw ───────────────────────────────────────
raw_path = Path(RAW_FILE)
assert raw_path.exists(), f"❌ Không tìm thấy file: {RAW_FILE}"

df = pd.read_csv(raw_path, encoding="utf-8")

print(f"✅ Đọc thành công: {df.shape[0]} dòng × {df.shape[1]} cột")
print(f"\n📋 Danh sách cột:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i:>2}. {col}")
```

```python
# ── Quick Preview ──────────────────────────────────────
df.head(3)
```

```python
# ── Lưu ra responses.csv ──────────────────────────────
df.to_csv(OUTPUT_FILE, index=False, encoding="utf-8-sig")

# ── Verify ────────────────────────────────────────────
verify = pd.read_csv(OUTPUT_FILE, encoding="utf-8-sig")
assert verify.shape == df.shape, "❌ File output không khớp shape!"
print(f"✅ Đã lưu thành công: {OUTPUT_FILE}")
print(f"   Shape: {verify.shape[0]} dòng × {verify.shape[1]} cột")
```
